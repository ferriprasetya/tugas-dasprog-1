# 2.Predict megawatts of Hydroelectric dam

```python
# height of the dam in meters
height = float(input('Enter height of the dam (in meters) => '))

# get flow rates in m^2/s
flow_rates = float(input('Enter flow rates (in m^3/s) => '))

density = 1000
megawatt = 10**6
efficiency = 0.9
gravity = 9.8
mass = flow_rates *  density

# w = mgh
potential_energy = mass * gravity * height
power_output = potential_energy * efficiency

# convert power output to megawatts
power_output /= megawatt

print(f'Power output of the dam is {power_output:.2f} MW')
```

# 3.Estimate temperature in freezer

```python
time = list(map(float, input('How long the freezer start since power failure with format hours minute => ').split()))
hours = time[0]
minutes = time[1]

# convert hours and minutes into hours
elapsed_time = hours + round(minutes / 60, 1)

# T = ((4t^2) / (t + 2)) - 20
temperature = ((4 * elapsed_time**2) / (elapsed_time + 2)) - 20

print(f'Freezer temperature is {temperature:.2f} Â°C')
```

# 6.Predict final exam score to passing desired grade

```python
desired_grade = input('Enter desired grade> ')
minimum_average = float(input('Enter minimum average required> '))
current_average = float(input('Enter current average in course> '))
final_percentage = float(input('Enter how much the final counts as a percentage of the course grade> '))

# conver final percentage into /100
final_percentage /= 100

# get weight of current average by the percentage
current_average = abs(current_average * (final_percentage-1))

# get needed score to reach minimum average
needed_final_score = minimum_average - current_average
needed_final_score /= final_percentage

print(f'You need a score of {needed_final_score:.2f} on the final to get a {desired_grade}')
```

# 7.Calculates how many BTU's of heat are delivered to a house

```python
total_gallons = int(input('Enter number of gallons of oil burned => '))
efficiency = float(input("Enter efficiency of the house's oil furnance (in percent) => "))

# 1 barrel = 42 gallons
total_barrel = total_gallons / 42

# 1 barrel =  5800000 BTU
percentage_efficiency = efficiency / 100
btu_per_barrel = 5800000
total_BTU = total_barrel * btu_per_barrel * percentage_efficiency

print('Total BTU delivered:   ', round(total_BTU, 2), 'BTU')
```

# 8.Calculate cost of water saved

```python
population = int(input('Enter total cummunity population => '))

total_toilets = round(population / 3)
avg_flush_per_day = 14
exiting_liter_per_flush = 15
existing_magnitude = total_toilets * (avg_flush_per_day * exiting_liter_per_flush)

new_liter_per_flush = 2
new_magnitude = total_toilets * (avg_flush_per_day * new_liter_per_flush)

cost_per_toilets = 150
new_toilet_cost = total_toilets * cost_per_toilets

magnitude_comparation = existing_magnitude - new_magnitude

print('Old magnitude is: ', existing_magnitude, 'liters/day')
print('New magnitude is: ', new_magnitude, 'liters/day')
print('The water saved after installing new toilet is: ',magnitude_comparation, 'liters/day' )
print('Cost of new toilets is: $', new_toilet_cost)

```

# 9.Calculates time required to cut the grass of yard

```python
yard_length = float(input('Enter your yard length (in feet) => '))
yard_width = float(input('Enter your yard width (in feet) => '))
house_length = float(input('Enter length of your house situated in the yard (in feet) => '))
house_width = float(input('Enter width of your house situated in the yard (in feet) => '))

# Calculate area
yard_area = yard_length * yard_width
house_area = house_length * house_width
grass_area = yard_area - house_area

# Calculate time to cut the grass of yard
# rate is 2 feet^2/s
rate = 2
time = round(grass_area / rate)
print(f'Time to cut the grass in the yard is {time} Seconds')
```

# 10.Calculate math equation of perpendicular bisector

```python
# input x1,x2,y1,y2
# gradien dari input
# x & y tengah
# gradien tegak lurus (-1 / gradien1)
# y = mx + b, b = y(tengah) - gradien2 . x(tengah)
# output: 2 titik & hasil persamaan

x1 = int(input('Enter x1 => '))
x2 = int(input('Enter x2 => '))
y1 = int(input('Enter y1 => '))
y2 = int(input('Enter y2 => '))
gradient = (y2 - y1) / (x2 - x1)

x_perpendicular = (x1 + x2) / 2
y_perpendicular = (y1 + y2) / 2
gradient_perpendicular = (-1 / gradient)
constanta = y_perpendicular - (gradient_perpendicular * x_perpendicular)

print(f'X1: {x1}, Y1: {y1}')
print(f'X2: {x2}, Y2: {y2}')
print(f'math equation of the perpendicular bisector: {constanta}')
```

# 11.Generating pythagorean triple by 2 given sides

```python
input_side1 = int(input('Enter 1st side of triangle => '))
input_side2 = int(input('Enter 2nd side of triangle (must be higher than 1st side) =>'))

side1 = input_side2**2 - input_side1**2
side2 = 2 * input_side1 * input_side2
hypotenuse = input_side1**2 + input_side2**2

print(f'Pythagorean triple is {side1}, {side2}, {hypotenuse}')
```
